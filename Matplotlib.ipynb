```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = ['A','B','C']
plt.plot(x,y)
plt.show()
```

```python
plt.title('Simple line plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

plt.plot(x,y)
plt.show()
```

```python
## Object Oriented Plot
## Fig : Entire figure, ax = Individual plot

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('X - AXIS')
ax.set_ylabel('Y - AXIS')
ax.set_title('Object Oriented')
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,20,25,30,40]
y2 = [30,25,5,10,44]

fig,ax = plt.subplots(1,2) # 1 Row, #2 Column

fig.suptitle('My Two Plots')

ax[0].plot(x, y,color='blue' )
ax[1].plot(x, y,color='red' )


ax[0].set_title('First Plot')
ax[1].set_title('Second Plot')

plt.show()
                
```

```python
fig = plt.figure(figsize=(3,2), dpi=200)
plt.plot(x,y)
plt.title("Customize Size & DPI")
plt.show()
```

```python
plt.plot(x,y)
plt.title('Save this Plot to Local')
plt.savefig('First Plot.png', dpi=300)
plt.close()
```

```python
## Legend

plt.plot(x,y1,label='Growth')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time vs Value')
plt.legend()
plt.show
```

```python
plt.plot(x,y1, color='green',linewidth=2, linestyle='--')
plt.show()
```

```python
plt.plot(x,y1, color='purple',linewidth=2, linestyle='--',marker='o',markersize=10)
plt.show()
```

```python
plt.plot(x,y1)
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()
```

```python


plt.plot(x,y1, color='purple',linewidth=2, linestyle='',marker='^',markersize=10)
plt.show()
```

```python
# Bar graphs
plt.bar(['A','B','C','D'],[5,7,3,10],color=['skyblue','lightgreen','salmon','red'],hatch=['/','\\','o','\\'])
plt.title('Bar Chart')
plt.show()
```

```python
# Scatter plot
plt.scatter(x,y,color='red')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram
Data = [1,2,2,3,3,3,4,4,4,4,5,5]
plt.hist(Data,bins=5)
plt.title('Histogram')
plt.show()
```

```python
# Pie chart
sizes =[30,45,25]
labels=['A','B','C']
colors=['#d689f5','#d4b729','#89f5ab']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
plt.show()
```
